```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
city_data = pd.read_csv("city_attributes.csv")
humidity_data = pd.read_csv("clean_humidity.csv")
temperature_data = pd.read_csv("clean_temperature.csv", index_col="datetime")
pressure_data = pd.read_csv("clean_pressure.csv", index_col="datetime")
wind_direction_data = pd.read_csv("clean_wind_direction.csv", index_col="datetime")
wind_speed_data = pd.read_csv("clean_wind_speed.csv", index_col="datetime")
```

```python
mult_city_data = pd.DataFrame(np.repeat(city_data.values, humidity_data.shape[0], axis=0), columns=city_data.columns)
# datetime = pd.DataFrame(humidity_data.datetime.values * city_data.shape[0], columns=['datetime'])
datetime = pd.DataFrame()
for index in range(city_data.shape[0]):
    datetime = pd.concat([datetime ,humidity_data.datetime], ignore_index=True)
datetime = datetime.rename(columns={0: 'datetime'})
```

```python
df = pd.concat([mult_city_data, datetime], axis=1)
df['datetime']= pd.to_datetime(df['datetime'])
```

```python
df[['humidity', 'temperature', 'pressure', 'wind_direction', 'wind_speed']] = np.nan
```

```python
for city in city_data.City.values:
    df.loc[df['City'] == city, 'humidity'] = humidity_data[city].tolist()
    df.loc[df['City'] == city, 'temperature'] = temperature_data[city].tolist()
    df.loc[df['City'] == city, 'pressure'] = pressure_data[city].tolist()
    df.loc[df['City'] == city, 'wind_direction'] = wind_direction_data[city].tolist()
    df.loc[df['City'] == city, 'wind_speed'] = wind_speed_data[city].tolist()
```

```python
df.to_parquet("../feature_store_repo/feature_repo/data/weather_data.pq")
```

```python
for city in city_data.City.values:
    df.loc[df['City'] == city, ['City','datetime']].to_csv(f"{city}_entity.csv")
```
