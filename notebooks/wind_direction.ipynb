```python
def find_outliers(col):
    q1 = col.quantile(.25)
    q3 = col.quantile(.75)
    IQR = q3 - q1
    ll = q1 - (1.5*IQR)
    ul = q3 + (1.5*IQR)
    upper_outliers = col.max() >= ul
    lower_outliers = col.min() <= ll
    have_outliers = upper_outliers + lower_outliers
    

    return have_outliers

import numpy as np
citie_with_outliers = []
for col in raw_data.columns:
    if raw_data[col].dtype in ["int64","float64"]:
        if find_outliers(raw_data[col]):
            citie_with_outliers.append(col)
```

```python
q1 = raw_data[citie_with_outliers].quantile(.25)
q3 = raw_data[citie_with_outliers].quantile(.75)
IQR = q3 - q1
ll = q1 - (1.5*IQR)
ul = q3 + (1.5*IQR)
    
raw_data[raw_data[citie_with_outliers] > ul] = np.nan
raw_data[raw_data[citie_with_outliers] < ll] = np.nan
raw_data.fillna(np.mean(raw_data), inplace=True)
# raw_data.loc[:, citie_with_outliers] = data[citie_with_outliers].to_numpy().tolist()
```

```python
raw_data.to_csv("../data/clean_data/clean_wind_direction.csv")
```
